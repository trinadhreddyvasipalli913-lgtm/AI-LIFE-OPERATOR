```python
class TaskAssistant:
    def __init__(self):
        self.tasks=[]
    def add_task(self):
        task = input("Enter a task: ").strip()
        if task == "":
            print("Task cannot be empty. ")
            return
        self.tasks.append(task)
        print("Task is added successfully. ")

    def show_tasks(self):
        if len(self.tasks) == 0:
            print("No tasks to show. ")
            return
        print("\nYour tasks: ")

        for i,task in enumerate(self.tasks,start=1):
            print(f"{i}. {task}")
        print()
    
    def run(self):
        
        print("Welcome to task manager: ")
        print("Type: add task | show task | exit")
        while True:
            command=input("enter command: ").strip().lower()
            if command == "add task":
                self.add_task()
            elif command == "show task":
                self.show_tasks()
            elif command =="exit":
                print("goodbye!,see you at schedule time")
                break
            else:
                print("Invalid command. please try again...")
if __name__ =="__main__":
    assistant = TaskAssistant()
    assistant.run()
```

```python
import json
class TaskAssistant:
    def __init__(self):
        try:
            with open("tasks.json","r") as file:
                self.tasks=json.load(file)
        except FileNotFoundError:
            self.tasks=[]
    def save_tasks(self):
        with open("tasks.json","w") as file:
            json.dump(self.tasks,file)
    def add_tasks(self,task):
        if task=="":
            print("tasks cannot be empty.")
            return
        self.tasks.append(task)
        self.save_tasks()
        print("Task added successfully.")
    def show_tasks(self):
        if len(self.tasks)==0:
            print("No tasks to show.")
            return
        print("\nYour tasks: ")
        for i,task in enumerate(self.tasks,start=1):
            print(f"{i}.{task}")
        print()
    def run(self):
        print("Welcome to task manager")
        print("Commands:")
        print("add task <task name>")
        print("show task")
        print("exit")
        while True:
            command=input("Enter command: ").strip().lower()
            if command.startswith("add task"):
                task=command[8:].strip()
                self.add_tasks(task)
            elif command=="show task":
                self.show_tasks()
            elif command=="exit":
                print("Goodbye! see you at scheduled time.")
                break
            else:
                print("Invalid command.please try again...")
if __name__=="__main__":
    assistant=TaskAssistant()
    assistant.run()
```

```python
import json
import requests


class AILifeOperator:

    def __init__(self):

        try:
            with open("tasks.json", "r") as file:
                self.tasks = json.load(file)

        except FileNotFoundError:
            self.tasks = []

    def save_tasks(self):

        with open("tasks.json", "w") as file:
            json.dump(self.tasks, file)

    def add_task(self, task):

        self.tasks.append(task)

        self.save_tasks()

        print("Task saved.")

    def show_tasks(self):

        if len(self.tasks) == 0:
            print("No tasks available.")
            return

        print("\nYour Tasks:\n")

        for i, task in enumerate(self.tasks, start=1):
            print(f"{i}. {task}")

    def ask_ai(self, user_input):

        prompt = f"""
        You are my personal AI Life Operator.

        These are my current tasks:

        {self.tasks}

        The user may ask:
        - create timetable
        - organize tasks
        - suggest priorities
        - plan the day
        - motivate
        - study planning
        - productivity advice

        Always use the task list when needed.

        User message:
        {user_input}
        """

        response = requests.post(
            "http://localhost:11434/api/generate",
            json={
                "model": "phi3",
                "prompt": prompt,
                "stream": False
            }
        )

        return response.json()["response"]

    def run(self):

        print("=== AI LIFE OPERATOR ===")

        print("\nCommands:")
        print("add task <task>")
        print("show task")
        print("exit")

        while True:

            user_input = input("\nYou: ")

            if user_input.lower() == "exit":

                print("Goodbye!")
                break

            elif user_input.lower().startswith("add task"):

                task = user_input[8:].strip()

                self.add_task(task)

            elif user_input.lower() == "show task":

                self.show_tasks()

            else:

                ai_reply = self.ask_ai(user_input)

                print("\nAI:\n")
                print(ai_reply)


if __name__ == "__main__":

    assistant = AILifeOperator()

    assistant.run()
```
